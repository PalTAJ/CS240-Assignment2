```python
def MakeNormalModel(weights):
    """Plots a CDF with a Normal model.

    """
    cdf = thinkstats2.Cdf(weights, label='Hits')

    mean, var = thinkstats2.TrimmedMeanVar(weights)
    std = np.sqrt(var)
    print('n, mean, std', len(weights), mean, std)

    xmin = 0
    xmax = mean + 4 * std

    xs, ps = thinkstats2.RenderNormalCdf(mean, std, xmin, xmax)
    thinkplot.Plot(xs, ps, label='model', linewidth=4, color='0.8')
    thinkplot.Cdf(cdf)
```

```python
"Here's the distribution of hits attempeted and a normal model, which is not a very good fit."


MakeNormalModel(data)
thinkplot.Config(title='Hits attempted, linear scale', xlabel='hits',
                 ylabel='CDF', loc='upper right')
```

```python
# Here's the distribution of attempted hits and a lognormal model, plotted on a log-x scale. The model is a better fit
#for the data.
log_d = np.log10(data)
MakeNormalModel(log_d)
thinkplot.Config(title='Hits Attempted, log scale', xlabel='Hits',
                 ylabel='CDF', loc='upper right')
```

```python
thinkplot.Scatter(data, data3, alpha=1,color='Orange')
thinkplot.Config(xlabel='Height (cm)',
                 ylabel='Weight (kg)',
                 legend=False)

#he data fall in obvious columns because they were rounded off.
# We can reduce this visual artifact by adding some random noice to the data.
```

```python
def Jitter(values, jitter=0.5):
    n = len(values)
    return np.random.normal(0, jitter, n) + values
dataa = Jitter(data, 1.4)
data33 = Jitter(data3, 0.5)

thinkplot.Scatter(dataa, data33, alpha=1.0,color="green")
thinkplot.Config(xlabel='Hits attempted ',
                 ylabel='Points',
                 legend=False)
# The columns are gone, but now we have a different problem: saturation. Where there are many overlapping points, 
# the plot is not as dark as it should be, which means that the outliers are darker than they should be, which gives
# the impression that the data are more scattered than they actually are.

# This is a surprisingly common problem, even in papers published in peer-reviewed journals.
# We can usually solve the saturation problem by adjusting alpha and the size of the markers, s.
```

```python
thinkplot.Scatter(dataa, data33, alpha=0.25, s=10,color='black')
thinkplot.Config(xlabel='hits attempted',
                 ylabel='points',
                 legend=False)

# That's better. This version of the figure shows the location and shape of the distribution most accurately. 
# There are still some apparent columns and rows where, most likely, data have the their hits and points using
# rounded values. If that effect is important, this figure makes it apparent; if it is not important, we could use more
# aggressive jittering to minimize it.
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov

print Cov(data,data3) 
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
```

```python
dataf = (data, data3)
ht = DiffMeansPermute(dataf)
pvalue = ht.PValue()
print pvalue


ht.PlotCdf()
thinkplot.Show()
```
